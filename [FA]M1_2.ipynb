```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
cereals = pd.read_csv('cereals.csv')
```

**1. Create and output the contingency table of Manuf and Type.**

```python
# Contigency table of Manuf and Type
cereals_conTable = pd.crosstab(cereals['Manuf'], cereals['Type'], margins = True, margins_name = 'Total')
cereals_conTable
```

**2. Create a stacked bar graph of the Manuf variable with Type overlay.**

```python
# Stacked bar graph of Manuf variable with Type overlay
cereals_conTable = pd.crosstab(cereals['Manuf'], cereals['Type'])
cereals_conTable.plot(kind = 'bar', stacked = True)
plt.title('Stacked bar graph of Manuf variable with Type overlay')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.show()
```

**3. Create a normalized bar graph of the Manuf variable with Type overlay.**

```python
# Normalized bar graph of the Manuf variable with Type overlay.
cereals_conTable = pd.crosstab(cereals['Manuf'], cereals['Type'], normalize = 'index')
cereals_conTable.plot(kind = 'bar', stacked = True)
plt.title('Normalized bar graph of the Manuf variable with Type overlay.')
plt.xlabel('Manufacturer')
plt.ylabel('Normalized Count')
plt.show()
```

**4. Create a stacked histogram of Calories with Manuf overlay.**

```python
# Stacked histogram of Calories with Manuf overlay
sns.histplot(cereals, x = 'Calories', hue = 'Manuf', multiple = 'stack', bins = 'auto')
plt.title('Stacked histogram of Calories with Manuf overlay')
plt.xlabel('Calories')
plt.ylabel('Count')
plt.show()
```

**5. Create a normalized histogram of Calories with Manuf overlay.**

```python
# Normalized histogram of Calories with Manuf overlay
sns.histplot(cereals, x = 'Calories', hue = 'Manuf', multiple = 'stack', bins = 'auto', stat='probability', kde = True)
plt.title('Normalized histogram of Calories with Manuf overlay')
plt.xlabel('Calories')
plt.ylabel('Normalized Percentage')
plt.xlim(75,150)
plt.show()
```

**6. Bin the Calories variable using bins for 0–90, 90–110, and over 110 calories. Create a bar chart of the binned calories variable with Manuf overlay.**

```python
# Bar chart of Calories (binned) with Manuf overlay
cereals['Calories_binned'] = pd.cut(x = cereals['Calories'], bins = [0, 90, 110.01, np.inf], labels=["0-90", "90-110", "Over 110"])

cereals_calTable = pd.crosstab(cereals['Calories_binned'], cereals['Manuf'], normalize = 'index')
cereals_calTable.plot(kind = 'bar', stacked = True)
plt.title('Bar chart of Calories (binned) with Manuf overlay')
plt.xlabel('Calories')
plt.ylabel('Count')
plt.show()
```
